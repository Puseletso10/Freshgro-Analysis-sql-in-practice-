## ðŸ“… Section 2: Date & Time Functions
### ðŸ”¹ 5. What year did most orders occur in?
> #### ðŸ’¼ "In which year did FreshGro receive the highest number of customer orders?"

```python
%%sql

SELECT
    YEAR(order_date) AS order_year,
    COUNT(*) AS total_orders
FROM
    orders
WHERE
    order_date IS NOT NULL
GROUP BY
    order_year
ORDER BY
    total_orders DESC
LIMIT 1;
```

### Explanation
* YEAR(order_date): Extracts the year from each orders's date.
* COUNT(*) : Counts the number of orders per year.
* WHERE order_date IS NOT NULL: Ensures no missing dates interfere with the count.
* GROUP BY order_year: Groups orders by year.
* ORDER BY total_orders DESC: Sorts the years by total number of orders in descending order.
* LIMIT 1: Returns only 1 year with most orders.

### ðŸ”¹ 6. How many days since the earliest order?
> #### ðŸ’¼ "How many days have passed since the first recorded customer order?"

```python
%%sql

SELECT
    DATEDIFF(CURDATE(),MIN(order_date)) AS Date_since_earliest_order
FROM
    orders
WHERE
    order_date IS NOT NULL;
```

### Explanation
* MIN(order_date): Finds the earliest order date in the dataset.
* CURDATE(): Gets the current date from the system.
* DATEDIFF(CURDATE(),MIN(order_date)): Calculates the number of days between today and the earliest order date.
* WHERE order_date IS NOT NULL: Ensured valid dates are considered.
* AS Date_since_earliest_order: Labes the result column for clarity.

### ðŸ”¹ 7. Add 7 days to each order date to simulate delivery due date
> #### ðŸ’¼ "What is the expected delivery due date for each order, assuming a 7-day delivery timeline?"

* `AS Delivery_due_date:` Assigns a clear name to the new column.
* `WHERE order_date IS NOT NULL:` Ensures only valid dates are used foe calculation.

### ðŸ”¹ 16. Use CASE to classify cities into regions 
> #### ðŸ’¼ "How can we group customer cities into regional categories to analyze trends across West, East, and other parts of Africa?"

```python
%%sql

SELECT city,
        CASE
            WHEN city IN ("Lagos", "Abuja", "Ibadan") THEN "Southwest"
            WHEN city IN ("Kano", "Kaduna", "Sokoto") THEN "Northwest"
            WHEN city IN ("Enugu", "Owerri", "Onitsha") THEN "Southeast"
            ELSE "Other"
        END AS region
FROM orders
WHERE city IS NOT NULL;
```

### Explanation
* `CASE WHEN city IN (...) THEN ...:` Classifies each city into a predefined region based on its name.
* ` 'Southwest', 'Northwest', 'Southeast':` Region labels assigned to cities in each group.
*  `ELSE 'Other':` labels other to any other city that doest match any listed group.
* `AS region:` Labels the output column with the name region.
* `WHERE city IS NOT NULL:` Ensure we only classify non-null city names.

### ðŸ”¹ 17. Nested IF: Label order quality 
> #### ðŸ’¼ "How can we categorize delivery experiences into quality tiers to better understand customer satisfaction?"

```python
%%sql

SELECT delivery_rating,
    IF (delivery_rating >=4, 'Excellent',
        IF(delivery_rating>=2, 'Okay', 'Poor')) AS rating_label
FROM orders;
```

### Explanation
* `IF (delivery_rating >=4, 'Excellent':` If the delivery rating is 4 or 5 it's labeled "Excellent".
* `IF(delivery_rating>=2, 'Okay':` If the delivery rating is 3 or 2 it's labeled "Okay".
* `"Poor":` Defaul label for all other cases

### ðŸ”¹ 18. Nested CASE + IF: Create a CASE logic that labels orders as Excellent, Average, Needs Improvement.

> #### ðŸ’¼ "How can we classify the quality of each order based on customer satisfaction and order size, so we can prioritize follow-up or improvement efforts?"

```python
%%sql

SELECT order_id,
        delivery_rating,
        quantity,
        CASE
            WHEN delivery_rating >= 4 THEN 'Excellent'
            WHEN delivery_rating = 3 THEN 'Average'
            WHEN delivery_rating <=2 THEN
                IF(quantity > 3, 'Needs Improvement', 'Poor')
            ELSE 'Unknown'
        END AS order_quality
FROM orders
WHERE delivery_rating IS NOT NULL;
```

### Explanation
* `CASE:` Used to define main delivery rating tiers (Excellent, Average, etc)
* `WHEN delivery_rating >= 4 THEN 'Excellent':` Orders rated 4 or 5 are considered excellent.
* `WHEN delivery_rating = 3 THEN 'Average':` Orders rated 3 are average.
* `WHEN delivery_rating <= 2 THEN IF(quantity > 3, 'Needs Improvement', 'Poor'):` For low-rated orders, we check quantity:
      * IF quantity is more than 3, it 'Needs Improvement'
      * Otherwise it's 'Poor'.
* `ELSE 'Unknown':` Catches any unexpected or invalid case.
* `WHERE delivery_rating IS NOT NULL:` Ensures we don't include null values.

## ðŸ”„ Section 6: Window Function Simulations (Simple Context) 

### ðŸ§  What is a Window Function in SQL?
A window function performs a calculation across a set of rows that are related to the current row, without collapsing the result into a single row (as aggregate functions do).

`It allows you to:`

- Retain row-level detail while also adding rankings, running totals, averages, or differences across rows.



### ðŸ”¹ 19. Rank orders by total amount (ROW_NUMBER) 
> #### ðŸ’¼ "Which customers placed the highest-value orders, and how do their purchases rank compared to others?"

```python
%%sql

SELECT order_id,
        customer_name,
        total_amount,
        ROW_NUMBER() OVER(ORDER BY total_amount DESC) AS order_rank
FROM orders
WHERE total_amount IS NOT NULL;
```

### Explanation 
* ROW_NUMBER() OVER(ORDER BY total_amount DESC): Assigns a unique rank to each order, starting from 1 to the highers total_amount in descending order.
* AS order_rank : Labels the output column that shows the rank
* WHERE total_amount IS NOT NULL: Filters out rows where the total_amount is missing to ensure accurate ranking

## ðŸ’¡ Bonus Section: Advanced Queries with Business Relevance
### ðŸ”¹ 20. Which cities have an average delivery rating below 3? 

> #### ðŸ’¼ "Which cities have an average delivery rating below 3, indicating potential issues with delivery satisfaction?"

```python
%%sql

SELECT city,
        ROUND(AVG(delivery_rating),1) AS avg_delivery_rating
FROM orders
WHERE delivery_rating IS NOT NULL AND city IS NOT NULL
GROUP BY city
HAVING avg_delivery_rating < 3;
```

### Explanation
* `(AVG(delivery_rating):` Calculates the average delivery rating for each city.
* `ROUND(...,1):` Rounds average rating to 1 decimal place for easier interpretation.
* `GROUP BY city:` Aggregates the result per city.
* `HAVING avg_delivery_rating < 3:` Filters to show only cities with an averagerating below 3.
* `WHERE delivery_rating IS NOT NULL AND city IS NOT NULL:` Ensure only valid values are used in the calculation.

### ðŸ”¹ 21. Which products are underperforming based on both order frequency and total revenue? 

> #### ðŸ’¼  "Which products are underperforming in both sales volume and total revenue, and may need to be reevaluated or promoted?"


```python
%%sql
SELECT product_name,
        COUNT(*) AS order_count,
        SUM(total_amount) AS total_revenue
FROM orders
WHERE product_name IS NOT NULL AND total_amount IS NOT NULL
GROUP BY product_name
HAVING COUNT(*) < 5 AND SUM(total_amount) < 100;
```

### Explanation
* `COUNT(*) :` Calculates how many times each product has been ordered.
* `SUM(total_amount):` Calculates the total revenue generated by each product.
* `GROUP BY product_name:` Aggregates the metrics per product.
* `HAVING COUNT(*) < 5 AND SUM(total_amount) < 100:` Filters to show only products with low order frequency and low revenue.
* `WHERE product_name IS NOT NULL AND total_amount IS NOT NULL:` Ensures only valid values are used.

### ðŸ”¹ 22. Rank customers by total amount spent (with ties given same rank 
> #### ðŸ’¼ "Who are our top-spending customers, and how do they rank based on total revenue contributed â€” including tied ranks?"

```python
%%sql

SELECT customer_name,
        SUM(total_amount) AS total_spent,
        RANK() OVER(ORDER BY SUM(total_amount) DESC) AS customer_rank
FROM orders
GROUP BY
    customer_name;
```

### Explanation
* `SUM(total_amount):` Calculates the total amount each customer has spent across all their orders.
* `RANK() OVER(ORDER BY SUM(total_amount) DESC):` Assigns a rank to each customer based on their total spending, ties recieve the same rank
* `GROUP BY customer_name:` aggregates the data per customer.

### ðŸ”¹ 23. Identify peak ordering day of the week with revenue totals
> #### ðŸ’¼ "Which day of the week generates the highest number of orders and revenue, and how do other days compare?"

```python
%%sql

SELECT 
    DAYNAME(order_date) AS day_of_week,
    COUNT(*) AS order_count,
    SUM(total_amount) AS total_revenue
FROM orders
WHERE order_date IS NOT NULL AND total_amount IS NOT NULL
GROUP BY DAYNAME(order_date)
ORDER BY total_revenue DESC;
```

### Explanation
* `DAYNAME(order_date):` Extracts the name of the day from the order_date (e.g Monday, Tuesday)
* `COUNT(*):` Counts how many orders were places on each day.
* `SUM(total_amount):` Adds up the total revenue for each day.
* `GROUP BY DAYNAME(order_date):` Groups data by day of the week.
* `ORDER BY total_revenue DESC:` Sorts the results to show the day with the highest revenue first.
* `WHERE order_date IS NOT NULL AND total_amount IS NOT NULL:` Filters out incomplete records.
